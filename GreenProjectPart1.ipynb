```python
import pandas as pd
from datetime import datetime, timedelta
import yfinance as yf
import matplotlib.pyplot as plt

# List of tickers
tickers = ['ENPH', 'NEE', 'IBDRY', 'VWDRY', 'DNNGY', 'FSLR', 'RUN', 'PLUG', 'BE', 'ORA', 'SEDG', 'SPWR', 'BLDP', 'CSIQ']

# Date five years ago from today
date_five_years_ago = datetime.now() - timedelta(days=365*5)
date_five_years_ago = date_five_years_ago.strftime('%Y-%m-%d')

# Date today
date_today = datetime.now().strftime('%Y-%m-%d')

# Create a DataFrame to store the ROIs
roi_df = pd.DataFrame(columns=['Ticker', 'ROI'])

# For each ticker, fetch the initial and final prices and calculate the ROI
for ticker in tickers:
    # Fetch the historical market data
    hist = yf.Ticker(ticker).history(start=date_five_years_ago, end=date_today)

    # Calculate the initial and final prices
    initial_price = hist.iloc[0]['Close']
    final_price = hist.iloc[-1]['Close']

    # Calculate the ROI
    roi = ((final_price - initial_price) / initial_price) * 100

    # Append to the DataFrame
    roi_df = roi_df.append({'Ticker': ticker, 'ROI': roi}, ignore_index=True)

# Display the DataFrame
print(roi_df)

# Create a bar plot of ROI
plt.figure(figsize=(10, 6))
plt.bar(roi_df['Ticker'], roi_df['ROI'])
plt.xlabel('Company')
plt.ylabel('ROI (%)')
plt.title('ROI of Green Energy Companies over the Past 5 Years')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def stock_analysis(df, ticker, window=20):
    # Reset index so we can use 'Date' as a regular column
    df = df.reset_index()
    
    # Ensure 'Date' is in datetime format
    df['Date'] = pd.to_datetime(df['Date'])
    
    # Calculate daily returns for each stock
    df['Daily Return'] = df.groupby('Ticker')['Adj Close'].pct_change()

    # Analyze returns
    print(df[df['Ticker'] == ticker]['Daily Return'].describe())

    # Rolling Statistics: Moving Average and Standard Deviation
    df['Rolling Mean'] = df.groupby('Ticker')['Adj Close'].transform(lambda x: x.rolling(window).mean())
    df['Rolling Std'] = df.groupby('Ticker')['Adj Close'].transform(lambda x: x.rolling(window).std())

    # Volume Analysis
    df['Volume Shifted'] = df.groupby('Ticker')['Volume'].shift()
    df['Volume Change'] = (df['Volume'] - df['Volume Shifted']) / df['Volume Shifted']

    # Plot rolling statistics and volume change for a single stock (e.g., 'ENPH')
    df_stock = df[df['Ticker'] == ticker]

    fig, axs = plt.subplots(3, figsize=(14, 7))

    axs[0].plot(df_stock['Date'], df_stock['Rolling Mean'], label='Rolling Mean')
    axs[0].plot(df_stock['Date'], df_stock['Adj Close'], label='Adjusted Close', alpha=0.3)
    axs[0].set_title('Rolling Mean for ' + ticker)
    axs[0].legend()

    axs[1].plot(df_stock['Date'], df_stock['Rolling Std'], label='Rolling Std')
    axs[1].set_title('Rolling Std for ' + ticker)
    axs[1].legend()

    axs[2].bar(df_stock['Date'], df_stock['Volume Change'], label='Volume Change')
    axs[2].set_title('Volume Change for ' + ticker)
    axs[2].legend()

    plt.tight_layout()
    plt.show()

```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

def decompose_time_series(df, ticker):
    # Filter the dataframe for the given ticker
    df_stock = df[df['Ticker'] == ticker]
    
    # Decompose 'Close' prices
    result = seasonal_decompose(df_stock['Close'], model='multiplicative', period = 1)
    
    # Plot the decomposed time series
    result.plot()
    plt.show()
```

```python
import yfinance as yf

tickers = ['NEE', 'ENPH', 'IBDRY', 'VWDRY', 'DNNGY', 'FSLR', 'RUN', 'PLUG', 'BE', 'ORA', 'SPWR', 'BLDP', 'CSIQ','SEDG']

for ticker in tickers:
    ticker_obj = yf.Ticker(ticker)
    financials = ticker_obj.financials.T  # Transpose the dataframe

    if not financials.empty:  # Check if data is available
        # Calculate profitability ratios
        if 'Gross Profit' in financials and 'Total Revenue' in financials:
            gross_margin = financials['Gross Profit'] / financials['Total Revenue']
            print(f"{ticker} Gross Margin: \n{gross_margin}\n")

        if 'Operating Income' in financials and 'Total Revenue' in financials:
            operating_margin = financials['Operating Income'] / financials['Total Revenue']
            print(f"{ticker} Operating Margin: \n{operating_margin}\n")

        if 'Net Income Applicable To Common Shares' in financials and 'Total Revenue' in financials:
            net_profit_margin = financials['Net Income Applicable To Common Shares'] / financials['Total Revenue']
            print(f"{ticker} Net Profit Margin: \n{net_profit_margin}\n")

        if 'Net Income Applicable To Common Shares' in financials and 'Total Assets' in financials:
            roa = financials['Net Income Applicable To Common Shares'] / financials['Total Assets']
            print(f"{ticker} Return on Assets (ROA): \n{roa}\n")

        if 'Net Income Applicable To Common Shares' in financials and 'Total Stockholder Equity' in financials:
            roe = financials['Net Income Applicable To Common Shares'] / financials['Total Stockholder Equity']
            print(f"{ticker} Return on Equity (ROE): \n{roe}\n")

```

```python
tickers = ["NEE", "ENPH", "IBDRY", "VWDRY", "DNNGY", "FSLR", "RUN", "PLUG", "BE", "ORA", "SPWR", "BLDP", "CSIQ", "SEDG"]  # List of tickers

for ticker in tickers:
    ticker_obj = yf.Ticker(ticker)
    balance_sheet = ticker_obj.balance_sheet.T  # Transpose the dataframe
    income_statement = ticker_obj.financials.T  # Transpose the dataframe
    
    # Calculate ratios
    if not balance_sheet.empty and not income_statement.empty:  # Check if data is available
        
        # Liquidity Ratios
        if 'Total Current Assets' in balance_sheet and 'Total Current Liabilities' in balance_sheet and 'Inventory' in balance_sheet:
            current_ratio = balance_sheet['Total Current Assets'] / balance_sheet['Total Current Liabilities']
            print(f'{ticker} Current Ratio: \n{current_ratio}\n')
            
            quick_ratio = (balance_sheet['Total Current Assets'] - balance_sheet['Inventory']) / balance_sheet['Total Current Liabilities']
            print(f'{ticker} Quick Ratio: \n{quick_ratio}\n')
        
        # Solvency Ratios
        if 'Total Liab' in balance_sheet and 'Total Stockholder Equity' in balance_sheet:
            debt_to_equity_ratio = balance_sheet['Total Liab'] / balance_sheet['Total Stockholder Equity']
            print(f'{ticker} Debt-to-Equity Ratio (D/E): \n{debt_to_equity_ratio}\n')
            
            equity_ratio = balance_sheet['Total Stockholder Equity'] / balance_sheet['Total Assets']
            print(f'{ticker} Equity Ratio: \n{equity_ratio}\n')

        # Efficiency Ratios
        if 'Cost of Revenue' in income_statement and 'Inventory' in balance_sheet:
            avg_inventory = balance_sheet['Inventory'].mean()  # Assuming constant inventory for simplicity
            inventory_turnover = income_statement['Cost of Revenue'] / avg_inventory
            print(f'{ticker} Inventory Turnover: \n{inventory_turnover}\n')

        if 'Total Revenue' in income_statement and 'Total Assets' in balance_sheet:
            avg_total_assets = balance_sheet['Total Assets'].mean()  # Assuming constant total assets for simplicity
            asset_turnover = income_statement['Total Revenue'] / avg_total_assets
            print(f'{ticker} Asset Turnover: \n{asset_turnover}\n')
```

```python
import pandas as pd
import yfinance as yf

tickers = ["NEE", "ENPH", "IBDRY", "VWDRY", "DNNGY", "FSLR", "RUN", "PLUG", "BE", "ORA", "SPWR", "BLDP", "CSIQ", "SEDG"]

# Initialize an empty DataFrame to store the results
asset_turnover_results = pd.DataFrame()

for ticker in tickers:
    ticker_obj = yf.Ticker(ticker)
    balance_sheet = ticker_obj.balance_sheet.T
    income_statement = ticker_obj.financials.T

    if not balance_sheet.empty and not income_statement.empty:
        if 'Total Revenue' in income_statement and 'Total Assets' in balance_sheet:
            avg_total_assets = balance_sheet['Total Assets'].mean()
            asset_turnover = income_statement['Total Revenue'] / avg_total_assets

            # Create a DataFrame for each ticker and append to the main DataFrame
            asset_turnover_df = asset_turnover.to_frame().reset_index()
            asset_turnover_df.columns = ['Year', 'Asset Turnover']
            asset_turnover_df['Ticker'] = ticker

            asset_turnover_results = pd.concat([asset_turnover_results, asset_turnover_df], ignore_index=True)

# Reorder columns
asset_turnover_results = asset_turnover_results[['Ticker', 'Year', 'Asset Turnover']]

# Print the results
print(asset_turnover_results)
```

```python
import yfinance as yf
import pandas as pd

# List of tickers
tickers = ['NEE', 'ENPH', 'IBDRY', 'VWDRY', 'DNNGY', 'FSLR', 'RUN', 'PLUG', 'BE', 'ORA', 'SPWR', 'BLDP', 'CSIQ', 'SEDG']

# DataFrame to store results
market_prospect_results = pd.DataFrame(columns=['Ticker', 'Year', 'EPS', 'P/E', 'Dividend Yield'])

for ticker in tickers:
    company = yf.Ticker(ticker)
    income_statement = company.financials.T
    shares_outstanding = company.info['sharesOutstanding']
    market_price = company.info['previousClose'] 
    annual_dividends = company.info['dividendsPerShare'] if 'dividendsPerShare' in company.info else 0

    if 'Net Income' in income_statement:
        income_statement['EPS'] = income_statement['Net Income'] / shares_outstanding
        income_statement['P/E'] = market_price / income_statement['EPS']
        income_statement['Dividend Yield'] = annual_dividends / market_price if market_price else 0
        income_statement = income_statement.reset_index()
        income_statement = income_statement.rename(columns={'index': 'Year'})
        income_statement['Ticker'] = ticker
        income_statement = income_statement[['Ticker', 'Year', 'EPS', 'P/E', 'Dividend Yield']]
        
        market_prospect_results = market_prospect_results.append(income_statement)

# Reset the index
market_prospect_results = market_prospect_results.reset_index(drop=True)

print(market_prospect_results)
```

```python
import yfinance as yf

def get_current_ratios(tickers):
    current_ratios = {}
    for ticker in tickers:
        company = yf.Ticker(ticker)
        try:
            current_ratio = company.info['currentRatio']
            current_ratios[ticker] = current_ratio
        except KeyError:
            current_ratios[ticker] = 'N/A'
    return current_ratios

tickers = ["NEE", "ENPH", "IBDRY", "VWDRY", "DNNGY", "FSLR", "RUN", "PLUG", "BE", "ORA", "SPWR", "BLDP", "CSIQ", "SEDG"]  # replace with your chosen tickers
current_ratios = get_current_ratios(tickers)
for ticker, ratio in current_ratios.items():
    print(f"The current ratio for {ticker} is {ratio}")
```
